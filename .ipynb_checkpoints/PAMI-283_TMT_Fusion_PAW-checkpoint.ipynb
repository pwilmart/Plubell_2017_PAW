# Can we verify which channels are the pooled standards?
## (without knowing ahead of time...)
We will see if we can load in the raw TMT data and find the pairs of pooled standard channels de novo.

```r
# library imports
library(tidyverse)
library(stringr)
library(limma)
library(psych)
library(Matrix)
```

## Load the grouped protein file
This can be either the grouped protein file or the labeled file (for reading by the IRS script). We have inspected the file using Excel and determined that the main table has 4437 rows (not counting the header row).

```r
# load the raw data
# the table starts in Row 5
# there are extra lines at the bottom - need to stop before the end
data_import <- read_tsv("PAW_labeled_grouped_protein_summary_TMT_8.txt", 
                        skip = 4, n_max = 4437, guess_max = 4437)

# the "Filter" column flags contams and decoys
data_all <- filter(data_import, is.na(Filter))
```

```r
nrow(data_import); nrow(data_all)
```

## Extract the raw data for each TMT-plex

```r
# make a data frame for each TMT-plex (and drop zeros)
get_data <- function(df, substring) {
    # get the columns
    df <- df %>%
      select(starts_with("TotInt")) %>% 
      select(contains(substring))
    
    # drop rows with zeros
    df[apply(df, 1, function(row) all(row !=0 )), ]
}

# get each experiment and set column names to something easy to parse
A_raw <- get_data(data_all, "_A_")
colnames(A_raw) <- str_c(as.character(1:10), "_A")
B_raw <- get_data(data_all, "_B_")
colnames(B_raw) <- str_c(as.character(1:10), "_B")
C_raw <- get_data(data_all, "_C_")
colnames(C_raw) <- str_c(as.character(1:10), "_C")
D_raw <- get_data(data_all, "_D_")
colnames(D_raw) <- str_c(as.character(1:10), "_D")
```

## We can do a sample loading normalization to reduce variation
We do not have to have any knowledge of which channels are in what groups to correct for basic protocol variation (protein and peptide assays, digestion, labeling, SepPak steps). We can safely do this step with no prior knowledge of much of anything.

```r
SL_Norm <- function(df) {
    # This makes each channel sum to the average grand total
    # A sample loading normalization because total protein is the same
    norm_facs <- mean(c(colSums(df))) / colSums(df)
    print(cat(sprintf("%s - %0.3f\n", colnames(df), norm_facs)))
    df_sl  <- sweep(df, 2, norm_facs, FUN = "*")
}

# normalize the raw data
A_sl <- SL_Norm(A_raw)
B_sl <- SL_Norm(B_raw)
C_sl <- SL_Norm(C_raw)
D_sl <- SL_Norm(D_raw)
```

# Try to guess the standard channel pair for Experiment A
We have a few options for measuring sample-to-sample similarity. The multi-panel scatter plots are worth trying. The scatter plots will probably be more useful. The correlation coefficients are usually close to 1.0 for everything in TMT datasets.

```r
# multi-panel scatter plot grids from the psych package
# experiment A
pairs.panels(log10(A_sl), lm = TRUE, main = "Experiment A")
```

## Looks like channels 6 and 8 (of 10) are the best correlated
We really only have one likely pair.

## Try the MDS cluster plots
Maybe clustering will let us know which samples are similar. We will simplify the sample labels so the plots are easier to read.

```r
# check the clustering
temp <- log2(A_sl)
plotMDS(temp, main = "Experiment A")
```

## Clustering is kind of busy (and unexpected)
The pair (6, 8) are the closest, but the scatter plot was pretty obvious and the clustering is not so clear. Let's compare the candidate pair in larger scatter plot along with another close pair (4, 5).

```r
pairs.panels(log10(A_sl[c(4, 5, 6, 8)]), lm = TRUE, main = "Experiment A candidate")
```

## (6, 8) seems to be the pair
We can also try something like a coefficient of variance. The CV distributions are useful for seeing positive effects of normalizations. For the pair-wise situation, we can do the absolute value of the difference divided by the mean for each protein. Instead of tracking the median, maybe we can get better distinguishing power if we use the inter quartile range.

```r
iqr_diff <- function(df) {
    # make an empty matrix ncol-by-ncol
    results  <- matrix(0.0, ncol(df), ncol(df))
    
    for (i in 1:ncol(df)) {
        for (j in (i+1):ncol(df)) {
            if (j > ncol(df)) {
                break
            }
            med_diff <- 100 * abs(df[i] - df[j]) / rowMeans(df[c(i, j)])
            results[i, j] <- IQR(med_diff[[1]])
        }
    }
    results
}

print("Experiment A")
print(round(data.matrix(iqr_diff(A_sl)), 2))
```

```r
x <- iqr_diff(A_sl)
# from https://stackoverflow.com/questions/38664241/ranking-and-counting-matrix-elements-in-r
x_long <- summary(Matrix::Matrix(x, sparse = TRUE))
y <- x_long[order(-x_long[, 3], decreasing = TRUE),]
for (row in 1:4){
    i <- y[row, 1]
    j <- y[row, 2]
    print(c(i, j))
}
```

```r
?order
```

## We can use a grid layout
The lowest IQR of 8.3% is for (6, 8) and it kiind of stands out. 

## We can also look at the whole distribution of the standardized differences

```r
diff <- function(df, x, y) {
    diff <- 100 * abs(df[x] - df[y]) / rowMeans(df[c(x, y)])
}

print("Experiment A")
boxplot(data.frame(diff(A_sl, 4, 5), diff(A_sl, 6, 8), diff(A_sl, 9, 10)))
```

## The 3rd distribution is better
The third distribution for (6, 8) has the smallest median, a smaller IQR, and a more controlled high value outlier tail. **The correct answer is that channels 6 and 8 in Experiment A are the standards.**

# What worked best?
I think that the scatter plots are the best. When there are too many plots and they are small, it is harder to pick the best one. If a few possibilities emerge, then larger plots with those candidates makes a pretty big difference. 

The cluster plots were okay, but more for picking the set of candidates that finding the best pair. The grids of standardized differences were kind of like the cluster plots. They help to see which pairs are the likely ones to explore further. The distributions of the differences as box plots were not too bad. They seem a little less effective compared to the scatter plots. It is hard to get a single computed value that definitively says these are the two pooled standard channels. The scatter plot shape is pretty distinctive for the standards. It is not just its tightness to the diagonal, though. It should be very tight at larger intensities and gently widen as the intensities get smaller. There should also be no (or very few) outlier points.
